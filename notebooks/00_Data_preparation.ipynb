Para este proyecto vamos a utilizar los datos del dataset de Yelp. En él hemos encontrado 5 ficheros distintos con formato JSON: yelp_academic_dataset_business, yelp_academic_dataset_checkin, yelp_academic_dataset_review, yelp_academic_dataset_tip, yelp_academic_dataset_user. Estos ficheros los debemos procesar para quedarnos solo con la información que nos interesa. Los pasos que vamos a seguir son los siguientes:

```python
# Libraries
import networkx as nx
import matplotlib.pyplot as plt
import json
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

## 1. Lectura de los ficheros.

En primer lugar, vamos a cargar los ficheros para ver la información que nos aportada cada uno y a partir de ellos decidiremos qué caso de uso queremos plantear. Hay algunos ficheros que hemos leído sin problemas, pero otros los hemos tenido que dividir debido a su tamaño.

### Fichero business

```python
# Reading yelp_academic_dataset_business.json
f = '../data/raw/yelp_academic_dataset_business.json'
data = open(f, 'r', encoding = 'utf-8').readlines()
bs = []
for ele in data:
    bs.append(json.loads(ele))
    
# We convert it to df because as a result of reading we have obtained a list
business = pd.DataFrame(bs)

business.head()
```

### Fichero checkin

```python
# Reading yelp_academic_dataset_checkin.json
f = '../data/raw/yelp_academic_dataset_checkin.json'
data = open(f, 'r', encoding = 'utf-8').readlines()
checkin = []
for ele in data:
    checkin.append(json.loads(ele))
    
# We convert it to df because as a result of reading we have obtained a list
checkin_df = pd.DataFrame(checkin)

checkin_df.head()
```

### Fichero tip 

```python
# Reading yelp_academic_dataset_tip.json
f = '../data/raw/yelp_academic_dataset_tip.json'
data = open(f, 'r', encoding = 'utf-8').readlines()
tip = []
for ele in data:
    tip.append(json.loads(ele))

# We convert it to df because as a result of reading we have obtained a list
tip_df = pd.DataFrame(tip)

tip_df.head()
```

### Fichero review 

```python
# We read yelp_academic_dataset_review.json. Because of the high number of data in review we obtained a memory error, so we are 
# going to separate the data in two parts. We have divided it by the date column. Firstly, we read rows with dates less 
# than 01-12-2017.
review1 = []

# We stablished lines = True and chunksize so that it loads when it reaches the chunksize limit.
with open("../data/raw/yelp_academic_dataset_review.json", "r", encoding = 'utf-8') as f:
    rev = pd.read_json(f, orient="records", lines=True, chunksize=1000)
# We drop review_id and text columns and stablished the limit date        
    for chunk in rev:
        reduced_chunk = chunk.drop(columns=['review_id', 'text'])\
                             .query("`date` <= '2017-12-01'")
        review1.append(reduced_chunk)

review1 = pd.concat(review1, ignore_index=True)

review1.head()
```

```python
#Second, we read the rows with date greater than 01-12-2017
review2 = []

with open("../data/raw/yelp_academic_dataset_review.json", "r", encoding = 'utf-8') as f:
    rev = pd.read_json(f, orient="records", lines=True, chunksize=1000)
        
    for chunk in rev:
        reduced_chunk = chunk.drop(columns=['review_id', 'text'])\
                             .query("`date` > '2017-12-01'")
        review2.append(reduced_chunk)
        
review2 = pd.concat(review2, ignore_index=True)

review2.head()
```

### Fichero user

```python
# We read yelp_academic_dataset_user.json. Because of the high number of data in review we obtained a memory error, so we 
# are going to separate the data in two parts. We have divided it by the date column. Firstly, we read rows with dates less 
# than 01-12-2017.
user = []

with open("../data/raw/yelp_academic_dataset_user.json", "r", encoding = 'utf-8') as f:
    rev = pd.read_json(f, orient="records", lines=True, chunksize=1000)
# We drop columns thar we are not going to use, to reduce the size of the data:        
    for chunk in rev:
        reduced_chunk = chunk.drop(columns=['name', 'useful', 'funny', 'cool', 'elite', 'friends', 'fans', 
                                            'compliment_more', 'compliment_profile', 'compliment_cute', 'compliment_list', 
                                            'compliment_note', 'compliment_plain', 'compliment_cool', 'compliment_funny', 
                                            'compliment_writer', 'compliment_photos'])\
                             .query("`yelping_since` <= '2017-12-01'")
        user.append(reduced_chunk)
        
user = pd.concat(user, ignore_index=True)

user.head()
```

```python
# Second, we read the rows with date greater than 01-12-2017
user2 = []

with open("../data/raw/yelp_academic_dataset_user.json", "r", encoding = 'utf-8') as f:
    rev = pd.read_json(f, orient="records", lines=True, chunksize=1000)
        
    for chunk in rev:
        reduced_chunk = chunk.drop(columns=['name', 'useful', 'funny', 'cool', 'elite', 'friends', 'fans', 
                                            'compliment_more', 'compliment_profile', 'compliment_cute', 'compliment_list', 
                                            'compliment_note', 'compliment_plain', 'compliment_cool', 'compliment_funny', 
                                            'compliment_writer', 'compliment_photos'])\
                             .query("`yelping_since` > '2017-12-01'")
        user2.append(reduced_chunk)
        
user2 = pd.concat(user2, ignore_index=True)

user2.head()
```

## 2. Decisión del caso de uso.

```python
fig, ax = plt.subplots()

# Create an empty graph
G = nx.Graph()

# Add some vertices
G.add_node('business')
G.add_node('user')
G.add_node('reviews')
G.add_node('tip')
G.add_node('checkin')
G.add_node('business_id')
G.add_node('user_id')

# Add some edges
G.add_edge('user', 'user_id')
G.add_edge('business', 'business_id')
G.add_edge('reviews', 'business_id')
G.add_edge('reviews', 'user_id')
G.add_edge('checkin', 'business_id')
G.add_edge('tip', 'user_id')
G.add_edge('tip', 'business_id')

# Draw the graph
nx.draw(G, with_labels=True, font_weight='bold', ax=ax)
```

Tras cargar los datos y tener una primera vista de ellos, ya podemos decidir el caso de uso que queremos plantear. En el grafo podemos observar los distintos archivos (User, Business, Reviews, Tip y Checkin) y sus relaciones, que están unidos por las variables *user_id* y *business_id*. 

- Business: Información sobre los negocios.
- Checkin: Aperture de los negocios.
- Tip: Comentarios de los usuarios sobre un negocio.
- Reviews: Reviews de los usuarios sobre un negocio.
- User: Información sobre los usuarios.

A partir de estos datos, el proyecto que queremos realizar es el siguiente:

**Probabilidad que tiene un restaurante de recibir una review con puntuación alta por parte un usuario experimentado, considerando:**

**- Puntuación alta: 4 o 5 estrellas.**

**- Usuario experientado: usuario que haya publicado más de 100 reseñas.**

Para ello vamos a emplear los ficheros *Business, reviews y user*, el resto no nos aporta información que nos sea útil para nuestro análisis. A partir de los ficheros seleccionados vamos a aplicar una serie de filtros y transformaciones para construir un solo fichero que contenga todos los datos necesarios para el caso.

## 3. Transformación y filtrado de los datos. 

A continuación, vamos a llevar a cabo una serie de transformaciones en los ficheros anteriormente mencionados de cara a prepararlos para nuestro análisis.

### Fichero business 

Este fichero como ya hemos mencionado contiene información sobre los establecimientos. Las columnas que principalmente nos interesan son las de categorias, para poder filtrar por restaurante, dado que hay negocios de todo tipo y la de atributos que contiene información sobre ciertas caracteristicas que se dan en un negocio.

La columna *attributes* contiene un diccionario, por lo que la tenemos que tratar para construir una columna para cada atributo.

```python
# We are going to use only the restaurants, so we filter the data by the category column, with some words that include 
# these businesses.
business = business[business['categories'].str.contains(
    'Restaurants|Food|Bars|Breakfast_brunch|Coffee|Sandwiches|Pizza|Burgers|Tea|Food Trucks|Mexican', 
    case=False, na=False
)]
```

```python
# We filter the data by the column is_open which means 1 if the restaurant is open. We are not interested in data from 
# closed establishments.
business = business[business['is_open'] == 1]

# We drop the columns that we are not going to use.
business = business.drop(['name', 'address', 'city', 'postal_code', 'stars', 'review_count', 'hours', 'is_open', 
                          'latitude', 'longitude', 'categories'], axis=1)

# We convert the attributes column into a different columns because we want to use them as variables for our models.
business_df = pd.concat([business.drop(['attributes'], axis=1), business['attributes'].apply(pd.Series)], axis=1)

business_df.head()
```

Vamos a analizar los valores nulos, ya que al haber construido distintas columnas con la variable *attributes* hemos observado que algunas de las columnas tienen nulos y queremos descartar que sean practicamente 100% nulas.

```python
# Analysing the missing values
nulos = business_df.isnull().sum()

nulos_porc = nulos/len(business_df) * 100

nulos_porc
```

En principio, vamos a eliminar las variables que tienen más de un 75% de nulos. Posteriormente, en el análisis EDA decidiremos si queremos deshacernos de alguna más.

```python
# We create a list with the variables that have less than 75% of missing values and we delete somo variables that we are
# not going to use.
variables_def = nulos_porc[nulos_porc < 75].drop(['BusinessParking', 'Ambience', 'GoodForMeal']).index

# We filter by the list that we have created so we elimite the columns that have more than 75% of missing values and the
# others columns that we are not going to use.
business_df = business_df[variables_def]

business_df.head()
```

Finalmente nos hemos quedado con un dataframe con el *business_id* para poder concatenar con el resto de ficheros, el estado en el que está el restaurante y una columna para cada atributo que puede tener un negocio.

### Fichero review 

Este archivo nos interesa sobre todo por la valoración de la review, es decir, el número de estrellas que le han dado al restaurante. También nos interesan el resto de sus variables *useful, funny y cool* que indican la valoración que otros usuarios han dado a la review.

```python
# We drop the date column because we only need it to separate the data
review1 = review1.drop('date', axis = 1)
review2 = review2.drop('date', axis = 1)

# We concatenate the two dataframes.
review_df = pd.concat([review1, review2], axis = 0)

review_df.head()
```

No hemos realizado ninguna transformación importante, hemos concatenado los df ya que tuvimos que separar el fichero para su lectura y eliminado la columna *date*. Nos quedamos con *user_id* y *business_id* ya que nos servirán para concatenar y construir el df final.

### Fichero user

La importancia de este fichero es principalmente para realizar el filtrado de datos por los usuarios experimentados, es decir, aquellos que han realizado más de 100 reseñas.

```python
# We drop the yelping_since column because we only need it to separate the data. We eliminate the columns average_stars 
# and compliment_hot too.
user1= user.drop(['yelping_since', 'average_stars', 'compliment_hot'], axis = 1)
user2 = user2.drop(['yelping_since', 'average_stars', 'compliment_hot'], axis = 1)

# We concatenate the two dataframes.
user_df = pd.concat([user1, user2], axis = 0)

user_df.head()
```

```python
# We filter the data to only have users with more than 100 reviews
user_df = user_df[user_df['review_count'] >= 100]
```

### FICHERO FINAL 

Por último, unimos los tres data frames que hemos transformado, para obtener el fichero final que utilizaremos en el resto del trabajo.

```python
# Finally, we merge the dataframes into one
df_final = pd.merge(business_df, review_df, on = 'business_id')
df_final = pd.merge(df_final, user_df, on = 'user_id')

df_final.head()
```

## 4. Exportación del fichero final.

```python
# We save the data in a csv file
df_final.to_csv('../data/raw/reviews_final.csv', index = False)
```
