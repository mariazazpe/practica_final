En este notebook vamos a realizar un modelo LightGBM. Los pasos que vamos a seguir son:

1. Importación de librerías, funciones y datos.
2. Creación del modelo.

4. Predicciones y evaluación.
5. Exportación del modelo.

## 1. Importación de librerías, funciones y datos.

```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import scikitplot as skplt
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.metrics import ConfusionMatrixDisplay, fbeta_score, plot_confusion_matrix
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, accuracy_score, classification_report
from sklearn.pipeline import Pipeline
import lightgbm as lgb
from sklearn.inspection import PartialDependenceDisplay
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
# Function to evaluate the model
def evaluate_model(ytest, ypred, ypred_proba = None):
    if ypred_proba is not None:
        print('ROC-AUC score of the model: {}'.format(roc_auc_score(ytest, ypred_proba[:, 1])))
    print('Accuracy of the model: {}\n'.format(accuracy_score(ytest, ypred)))
    print('Classification report: \n{}\n'.format(classification_report(ytest, ypred)))
```

```python
# Load data
X_train_val = pd.read_csv("../data/processed/X_train_val.csv",index_col=0 )
y_train_val = pd.read_csv("../data/processed/y_train_val.csv",index_col=0 )

x_val = pd.read_csv("../data/processed/X_val.csv",index_col=0 )
y_val = pd.read_csv("../data/processed/y_val.csv",index_col=0 )
```

## 2. Creación del modelo.

```python
# Import the preprocessor.
preprocessor = pickle.load(open('../models/preprocessor.pickle', 'rb'))

# Classifier.
classifier = lgb.LGBMClassifier(n_jobs = -1, random_state = 12345, objective='Binary')
```

```python
lightgbm_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', classifier)
])
```

## 3. Entrenamiento del modelo.

```python
lightgbm_model.fit(X_train_val, y_train_val)
```

## 4. Predicciones y evaluación.

```python
# Predictions of target value for the validation set.
pred = lightgbm_model.predict(x_val)

# Predictions of probabilities of the target value for the validation set.
pred_proba = lightgbm_model.predict_proba(x_val)
```

### Precision, Recall, F1 y accuracy.

```python
evaluate_model(y_val, pred)
```

- **Support** nos dice los datos reales que tenemos, 43264 que son 0 y 110766 que son 1.


- La **precisión** se calcula con las columnas de la matriz de confusión y su precisión es 0.63 para los casos 0 y 0.75 para los casos 1. Tiene una predicción similar para los 0 y 1.


- El **recall** se calcula con las filas de la matriz de confusión, la predicción no es muy buena porque no logra identificar las puntuaciones bajas, aunque acierta practicamente todos los 1 (0.96).


- El **f1** es la combinación de las dos medidas anteriores, nos permite valorar la predicción con más exactitud. Identifica 0.84 de puntuación alta y 0.27 de puntuación baja.


- El **accuracy** se calcula con la matriz de confusión y es 0.74, es decir, que el modelo predice correctamente el 74% de los casos.

### Matriz de confusión. 

```python
titles_options = [("Confusion matrix, without normalization", None),
                  ("Normalized confusion matrix", 'true')]

for title, normalize in titles_options:
    disp = plot_confusion_matrix(lightgbm_model, x_val, y_val,
                                 cmap=plt.cm.Blues,
                                 normalize=normalize)
    disp.ax_.set_title(title)
    print(title)
    print(disp.confusion_matrix)

plt.show()
```

La matriz de confusión muestra que el modelo acierta 7468 casos que son 0 y los clasifica de esa manera, pero falla 35796 que clasifica como 1 y son 0. En el caso de los 1, falla 4454 que clasifica como 0 y acierta 106312. El desempeño es muy bueno para la clase 1, pero falla muchos casos que son 0, más concretamente el 83% de ellos.

### Curva ROC.

```python
# Keep probabilities for the positive outcome only
yhat = pred_proba[:, 1]
# Calculate roc curves
fpr, tpr, thresholds = roc_curve(y_val, yhat)
# Plot the roc curve for the model
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='Light GBM')
# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
# Show the plot
plt.show()
```

La clave de la curva ROC es que cuanto más cerca esté de la esquina superior izquierda de nuestro gráfico y cuanta más área haya debajo de la curva, mejor será el modelo. En este caso se aprecia que la curva está por encima de la línea de puntos, lo que indica que tiene capacidad predictiva, pero no es un buen modelo, su área debajo de la curva no es muy amplia y hacía el final se comienza a apreciar una tendencia a la disminución, lo que indica que el modelo comienza a fallar más.

### Curva Lift.

```python
skplt.metrics.plot_lift_curve(y_val, pred_proba)
plt.show()
```

El modelo comienza prediciendo mejor la clase 0, pero rapidamente disminuye, la clase 1 más o menos se mantiene la predicción para todos los porcentajes de muestra, siendo mejor el modelo que la aletoriedad.

```python
skplt.metrics.plot_cumulative_gain(y_val, pred_proba)
plt.show()
```

La ganancia de la clase 0 es superior a la de la clase 1, tomando por ejemplo el 20% de la muestra, se detectaría sobre el 25% de la clase 1 y cerca del 40% de la clase 0. Esta probabilidad aumenta a media que se incrementa el tamaño de la muestra.
