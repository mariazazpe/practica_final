## Importamos librerias, funciones y datos

```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import scikitplot as skplt
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.metrics import ConfusionMatrixDisplay, fbeta_score, plot_confusion_matrix
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, accuracy_score, classification_report
from sklearn.pipeline import Pipeline
import lightgbm as lgb
from sklearn.inspection import PartialDependenceDisplay
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
#Function to evaluate the model
def evaluate_model(ytest, ypred, ypred_proba = None):
    if ypred_proba is not None:
        print('ROC-AUC score of the model: {}'.format(roc_auc_score(ytest, ypred_proba[:, 1])))
    print('Accuracy of the model: {}\n'.format(accuracy_score(ytest, ypred)))
    print('Classification report: \n{}\n'.format(classification_report(ytest, ypred)))
```

```python
# Load data
X_train_val = pd.read_csv("../data/processed/X_train_val.csv",index_col=0 )
y_train_val = pd.read_csv("../data/processed/y_train_val.csv",index_col=0 )

x_val = pd.read_csv("../data/processed/X_val.csv",index_col=0 )
y_val = pd.read_csv("../data/processed/y_val.csv",index_col=0 )
```

## Creamos el modelo

Creamos el modelo LightGBM con los datos de entrenamiento y validación.

```python
# Import the preprocessor.
preprocessor = pickle.load(open('../modelos/preprocessor.pickle', 'rb'))

#classifier
classifier = lgb.LGBMClassifier(n_jobs = -1, random_state = 12345, objective='Binary')
```

```python
lightgbm_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('clasificador', classifier)
])
```

## Optimizacion del modelo

```python
# Optimization of the light_gbm with the train set: input and target data. 
lightgbm_model.fit(X_train_val, y_train_val)
```

## Predicciones y evaluación del modelo

```python
pred = lightgbm_model.predict(x_val)

pred_proba = lightgbm_model.predict_proba(x_val)
```

```python
evaluate_model(y_val, pred)
```

El soporte nos dice los datos reales que tenemos, 48462 que son 0 y 125622 que son 1. 

La precisión es la evaluación de la segunda columna de la matriz de confusión y su precisión es 0.63, para los casos 0, pero es 0.75 (alta) para los casos 1. Identifica parecido los 0 y 1.

El recall se refiere a la segunda fila de la matriz de confusión, la predicción no es muy buena porque no logra identificar las no estrellas.

El f1 es la combinación de estas dos medidas y da el mismo resultado. Identifica 0.84 estrellas y 0.27 no estrellas.

El accuracy es la media de las dos columnas de la matriz de confusión y es 0.74, es decir, que el modelo predice correctamente el 74% de los casos.

## Matriz de confusión

```python
titles_options = [("Confusion matrix, without normalization", None),
                  ("Normalized confusion matrix", 'true')]

# Plot the confusion matrix
for title, normalize in titles_options:
    disp = plot_confusion_matrix(lightgbm_model, x_val, y_val,
                                 # display_labels=ytest,
                                 cmap=plt.cm.Blues,
                                 normalize=normalize)
    disp.ax_.set_title(title)
    print(title)
    print(disp.confusion_matrix)

plt.show()
```

La matriz de confusión muestra que el modelo tiene un buen desempeño, ya que la mayoría de los casos se clasifican correctamente. Sin embargo, el modelo tiene un desempeño mucho mejor para la clase 0 que para la clase 1. Esto se debe a que la clase 1 tiene menos casos que la clase 0. Por lo tanto, el modelo tiende a clasificar la mayoría de los casos como clase 0. Esto se puede ver en la matriz de confusión, donde la mayoría de los casos se clasifican como clase 0. Sin embargo, el modelo tiene un buen desempeño para la clase 1, ya que la mayoría de los casos se clasifican correctamente.

## Curva ROC

```python
# Keep probabilities for the positive outcome only
yhat = pred_proba[:, 1]
# Calculate roc curves
fpr, tpr, thresholds = roc_curve(y_val, yhat)
# Plot the roc curve for the model
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='Light GBM')
# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
# Show the plot
plt.show()
```

La curva ROC muestra que el modelo tiene un buen desempeño, ya que se encuentra por encima de la línea punteada, que representa el modelo que no tiene capacidad de predicción.

 ## Curva Lift

```python
skplt.metrics.plot_lift_curve(y_val, pred_proba)
plt.show()
```

La curva Lift muestra como el modelo predice mejor los casos que son 0 que en los casos que son 1. 

## Curva de ganancia

```python
skplt.metrics.plot_cumulative_gain(y_val, pred_proba)
plt.show()
```

La curva de ganancia muestra que el modelo tiene un buen desempeño, ya que la ganancia es positiva y se encuentra por encima de la línea punteada, que representa el modelo que no tiene capacidad de predicción.

## Gráfico de dependenciqa parcial

```python
features = [0, 1, (0, 1)]
PartialDependenceDisplay.from_estimator(lightgbm_model, x_val, features)
```

Este gráfico de dependencia parcial muestra la relacion entre estas dos variables, es alta.
