## Importamos librerias, funciones y datos

```python
#Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import pickle
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.metrics import ConfusionMatrixDisplay, fbeta_score, plot_confusion_matrix
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, accuracy_score, classification_report
from sklearn.pipeline import Pipeline
from sklearn.dummy import DummyClassifier
import scikitplot as skplt
from sklearn.inspection import PartialDependenceDisplay
import warnings
warnings.filterwarnings('ignore')
```

```python
#Function to evaluate the model
def evaluate_model(ytest, ypred, ypred_proba = None):
    if ypred_proba is not None:
        print('ROC-AUC score of the model: {}'.format(roc_auc_score(ytest, ypred_proba[:, 1])))
    print('Accuracy of the model: {}\n'.format(accuracy_score(ytest, ypred)))
    print('Classification report: \n{}\n'.format(classification_report(ytest, ypred)))
```

```python
# Load data
X_train_val = pd.read_csv("../data/processed/X_train_val.csv",index_col=0 )
y_train_val = pd.read_csv("../data/processed/y_train_val.csv",index_col=0 )

x_val = pd.read_csv("../data/processed/X_val.csv",index_col=0 )
y_val = pd.read_csv("../data/processed/y_val.csv",index_col=0 )
```

## Creamos el modelo

Vamos a utilizar el modelo DummyClassifier como modelo base.

```python
#Import the preprocessor.
preprocessor = pickle.load(open('../modelos/preprocessor.pickle', 'rb'))

#Classifier
classifier = DummyClassifier(strategy = 'most_frequent', random_state = 12345)
```

```python
# We create a pipeline with the preprocessor and the classifier.
dummy_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('clasificador',  classifier)
])
```

## Opmimizamos el modelo

Optimizamos el modelo con el train.

```python
# Optimization of the dummy_model with the train set: input and target data. 
dummy_model.fit(X_train_val, y_train_val)
```

## Predicciones y evaluación

Predecimos con la validación.

```python
# Predict the target value of the validation set.
pred = dummy_model.predict(x_val)

# Predict the probabilities of the target value for the validation set.
pred_proba = dummy_model.predict_proba(x_val)
```

```python
# We evaluate the model with differents metrics.
evaluate_model(y_val, pred)
```

El soporte nos dice los datos reales que tenemos, 48462 que son 0 y 125622 que son 1. 

La precisión es la evaluación de la segunda columna de la matriz de confusión y su precisión es 0 (muy baja) para los casos 0, pero es 0.72 (alta) para los casos 1, no logra identificar ninguna estrella.

El recall se refiere a la segunda fila de la matriz de confusión, la predicción es mala porque no logra identificar las no estrellas.

El f1 es la combinación de estas dos medidas y da el mismo resultado. Identifica 0.84 estrellas, pero no identifica ninguna no estrella.

La precisión es alta, pero debido a que la variable de destino está desequilibrada y la mayoría son 1. 

El accuracy es la media de las dos columnas de la matriz de confusión y es 0.72, es decir, que el modelo predice correctamente el 72% de los casos.

## Matriz de confusión

```python
titles_options = [("Confusion matrix, without normalization", None),
                  ("Normalized confusion matrix", 'true')]

# Loop to plot the confusion matrix normalized and not normalized.
for title, normalize in titles_options: 
    disp = plot_confusion_matrix(dummy_model, x_val, y_val,
                                 # display_labels=ytest,
                                 cmap=plt.cm.Blues,
                                 normalize=normalize)
    disp.ax_.set_title(title)
    print(title)
    print(disp.confusion_matrix)
    
#Display the confusion matrix.
plt.show()
```

Podemos ver como el modelo predice 48462 1 y ningún 0 en los casos 0, pero muy bien los casos 1.

## Curva ROC

```python
# Keep probabilities for the positive outcome only.
yhat = pred_proba[:, 1]

# Calculate roc curves.
fpr, tpr, thresholds = roc_curve(y_val, yhat)

# Plot the roc curve for the model.
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='Dummy Classifier')

# Axis labels.
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()

# Show the plot.
plt.show()
```

Nuestro modelo no tiene capacidad de predicción, por lo que la curva ROC es una línea recta.

## Curva Lift

```python
# Lift curve
skplt.metrics.plot_lift_curve(y_val, pred_proba)
plt.show()
```

Como vemos en la curva lift, el modelo no tiene capacidad de predicción, por lo que la curva lift es una línea recta.

## Curva de ganancia

```python
# Cumulative gains curve
skplt.metrics.plot_cumulative_gain(y_val, pred_proba)
plt.show()
```

La curva de ganancia es una línea recta, por lo que el modelo no tiene capacidad de predicción.
