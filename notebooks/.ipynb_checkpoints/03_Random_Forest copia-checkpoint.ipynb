## Importamos librerias, funciones y datos

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import pickle 
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.metrics import ConfusionMatrixDisplay, fbeta_score, plot_confusion_matrix
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, accuracy_score, classification_report
from imblearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
import scikitplot as skplt
from sklearn.inspection import PartialDependenceDisplay
import warnings
warnings.filterwarnings('ignore')
```

```python
#Function to evaluate the model
def evaluate_model(ytest, ypred, ypred_proba = None):
    if ypred_proba is not None:
        print('ROC-AUC score of the model: {}'.format(roc_auc_score(ytest, ypred_proba[:, 1])))
    print('Accuracy of the model: {}\n'.format(accuracy_score(ytest, ypred)))
    print('Classification report: \n{}\n'.format(classification_report(ytest, ypred)))
```

```python
# Load data
X_train_val = pd.read_csv("../data/processed/X_train_val.csv",index_col=0 )
y_train_val = pd.read_csv("../data/processed/y_train_val.csv",index_col=0 )

x_val = pd.read_csv("../data/processed/X_val.csv",index_col=0 )
y_val = pd.read_csv("../data/processed/y_val.csv",index_col=0 )
```

## Creamos el modelo

Vamos a utilizar el modelo RandomForest,

```python
# Import the preprocessor.
preprocessor = pickle.load(open('../modelos/preprocessor.pickle', 'rb'))

#Classifier
classifier = RandomForestClassifier(random_state = 12345, n_jobs = -1, class_weight = 'balanced')
```

```python
random_forest = Pipeline([
    ('preprocessor', preprocessor),
    ('clasificador', classifier)   
])
```

## Optimizamos el modelo

Se optimiza el modelo con el train

```python
# Optimization of the random forest model with the train set: input and target data. 
random_forest.fit(X_train_val, y_train_val)
```

## Prediccion y evaluacion del modelo

```python
# Predict target value of the validation set.
pred = random_forest.predict(x_val)

# Predict probabilities of target value for validation set.
pred_proba = random_forest.predict_proba(x_val)
```

```python
evaluate_model(y_val, pred)
```

El soporte nos dice los datos reales que tenemos, 48462 que son 0 y 125622 que son 1. 

La precisión es la evaluación de la segunda columna de la matriz de confusión y su precisión es 0.4, para los casos 0, pero es 0.79 (alta) para los casos 1. Identifica algún 0.

El recall se refiere a la segunda fila de la matriz de confusión, la predicción no es muy mala porque logra identificar las no estrellas.

El f1 es la combinación de estas dos medidas y da el mismo resultado. Identifica 0.74 estrellas y 0.46 no estrellas.

La precisión es alta, pero debido a que la variable de destino está desequilibrada y la mayoría son 1. 

El accuracy es la media de las dos columnas de la matriz de confusión y es 0.65, es decir, que el modelo predice correctamente el 65% de los casos.

## Matriz de confusión

```python
titles_options = [("Confusion matrix, without normalization", None),
                  ("Normalized confusion matrix", 'true')]

# Plot the confusion matrix
for title, normalize in titles_options:
    disp = plot_confusion_matrix(random_forest, x_val, y_val,
                                 # display_labels=ytest,
                                 cmap=plt.cm.Blues,
                                 normalize=normalize)
    disp.ax_.set_title(title)
    print(title)
    print(disp.confusion_matrix)

plt.show()
```

La matriz de confusión nos muestra que el modelo predice 22682 1 que deberian ser 0 y 25780 0 correctamente. Por otro lado, predice 38286 0 que deberian ser 1 y 87338 1 correctamente. 

## Curva ROC

```python
# Keep probabilities for the positive outcome only
yhat = pred_proba[:, 1]
# Calculate roc curves
fpr, tpr, thresholds = roc_curve(y_val, yhat)
# Plot the roc curve for the model
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='Random Forest')
# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
# Show the plot
plt.show()
```

La curva ROC nos muestra que el modelo es bueno para predecir las estrellas, pero no tanto para las no estrellas.

## Curva Lift

```python
skplt.metrics.plot_lift_curve(y_val, pred_proba)
plt.show()
```

La curva Lift muestra que el modelo es bueno para predecir las estrellas, pero no tanto para las no estrellas.

## Curva de ganancia

```python
skplt.metrics.plot_cumulative_gain(y_val, pred_proba)
plt.show()
```

La curva de ganancia muestra que el modelo es capaz de clasificar correctamente los 1 y peor los 0.

## Gráfico de dependencia parcial

```python
features = [0, 1, (0, 1)]
PartialDependenceDisplay.from_estimator(random_forest, x_val, features)
```

EL gráfico de dependencia parcial muestra la dependendia de las variables, es alta.
